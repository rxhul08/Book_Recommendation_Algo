```python
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('Title')['Rating'].count().reset_index()
```

```python
number_rating.rename(columns={'Rating':'Number of Ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating, on='Title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['Number of Ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User_ID','Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='User_ID', index='Title', values='Rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
#K-nearest neighbors clustering algorithm. If one user is choosing one book, show relevant ones
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(book_pivot.iloc[237, :].values.reshape(1,-1), n_neighbors=6)
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    # Check if the book name exists in the index
    if book_name in book_pivot.index:
        # Get the index position of the book name
        book_id = np.where(book_pivot.index == book_name)[0]
        # Check if the book ID is found
        if len(book_id) > 0:
            book_id = book_id[0]
            # Find recommendations for the book
            distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
            # Print suggestions
            print("The suggestions for", book_name, "are:")
            for i in range(len(suggestions)):
                print(book_pivot.index[suggestions[i]])
        else:
            print(f"No recommendations found for {book_name}.")
    else:
        print(f"The book '{book_name}' is not found in the dataset.")
```

```python
recommend_book('Animal Farm')
```
